```python
# importing necessary libraries for demonstrating power transformation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PowerTransformer
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import scipy.stats as stats
```

```python
# loading the data
df = pd.read_csv('concrete_data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
# splitting our data into features and labels
x=df.drop(columns=['Strength'])
y=df.iloc[:,-1]
```

```python
# applying train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# plotting  displot before transformation

for col in x_train.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(x_train[col])
    plt.title(col)

    plt.subplot(122)
    stats.probplot(x_train[col], dist="norm", plot=plt)
    plt.title(col)

    plt.show()
```

```python
# now applying Box-cox transformation to the x_train columns
pt= PowerTransformer(method='box-cox')
x_train_transformed=pt.fit_transform(x_train+0.00000001)
x_test_transformed=pt.transform(x_test+0.00000001)

# converting the data into dataframe
x_train_transformed=pd.DataFrame(x_train_transformed,columns=x_train.columns)
x_test_transformed=pd.DataFrame(x_test_transformed,columns=x_test.columns)
```

```python
## plotting  displot after transformation

for col in x_train_transformed.columns:
  plt.figure(figsize=(14,4))
  plt.subplot(121)
  sns.distplot(x_train[col])
  plt.title(col)

  plt.subplot(122)
  sns.distplot(x_train_transformed[col])
  plt.title(col)

  plt.show()
```

```python
x_train_transformed.describe()
```

```python
# now applying Yeo-Jhonson transformation to the x_train columns
pt= PowerTransformer()
x_train_transformed2=pt.fit_transform(x_train)
x_test_transformed2=pt.transform(x_test)

# converting the data into dataframe
x_train_transformed2=pd.DataFrame(x_train_transformed,columns=x_train.columns)
x_test_transformed2=pd.DataFrame(x_test_transformed,columns=x_test.columns)
```

```python
## plotting  displot after transformation

for col in x_train_transformed2.columns:
  plt.figure(figsize=(14,4))
  plt.subplot(121)
  sns.distplot(x_train[col])
  plt.title(col)

  plt.subplot(122)
  sns.distplot(x_train_transformed2[col])
  plt.title(col)

  plt.show()
```

```python
x_train_transformed2.describe()
```

```python
# applying linear regression
lr=LinearRegression()
lr.fit(x_train_transformed2,y_train)
y_pred=lr.predict(x_test_transformed2)
```

```python
# checking r2 score
r2_score(y_test,y_pred)
```

```python
# applying mean cross validation
pt = PowerTransformer()
X_transformed2 = pt.fit_transform(x)
np.mean(cross_val_score(lr,X_transformed2,y,scoring='r2'))
```
